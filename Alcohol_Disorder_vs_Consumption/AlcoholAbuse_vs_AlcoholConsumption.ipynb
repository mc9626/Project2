```python
#extract raw data from kaggle.com
csv_file = "share-with-alcohol-use-disorder-vs-alcohol-consumption.csv"
AB_data_df = pd.read_csv(csv_file)
AB_data_df.head()
```

```python
new_df = AB_data_df[AB_data_df.Year == '2016']
new_df.head()
```

```python
Renamed_df = new_df.rename(columns={'Share of population with alcohol use disorder (%)':'Alcohol_Use_Disorder','Alcohol consumption (litres per capita) (liters of pure alcohol, projected estimates, 15+ years of age)': 'Alcohol_Consumption'})
```

```python
Round_df = Renamed_df.round({'Alcohol_Use_Disorder': 2})
Final_df = Round_df.dropna()
Final_df.head()
```

```python
Final_df.to_csv('AlcoholAbuse.csv', index=False)
```
